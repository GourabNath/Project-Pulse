```python
# Generate variable description and variable type
import pandas as pd
import json
df = pd.read_csv("/Attrition/Employee_Attrition.csv")

problem = '''
The company has seen consistent growth over the past few years, supported by a dedicated workforce. However, internal reports have highlighted an unusual pattern:
each year, nearly one in five employees leaves the organization unexpectedly. This turnover affects team stability and productivity, yet it’s unclear what underlying
factors are driving such a significant rate of attrition. Despite efforts to improve employee engagement, the company continues to face challenges in retaining talent.
As it seeks to refine its human resources strategies, the key questions remain: what is causing this persistent departure trend, and how can it be effectively
addressed?
'''

from src.engines.utils.variable_description_helper import generate_variable_description, extract_metadata


dic = {"variable": [], "variable_description": [], "variable_type": [], "reasoning": []}

for variable in list(df):
    metadata = extract_metadata(df, variable)
    variable_description = generate_variable_description(variable, problem, metadata)
    json_to_dict = json.loads(variable_description)

    dic["variable"].append(variable)
    dic["variable_description"].append(json_to_dict["description"])
    dic["variable_type"].append(json_to_dict["variable_type"])
    dic["reasoning"].append(json_to_dict["reason"])

data_df = pd.DataFrame(dic)
```

```python
data_df.to_csv("/Attrition/variable_description.csv", index=False)
```
