## Generate Summary and Plots for a List of Numerical Variables

```python
numerical_variables = ['Age',
                       'DailyRate',
                       'DistanceFromHome',
                       'HourlyRate',
                       'MonthlyIncome',
                       'MonthlyRate',
                       'NumCompaniesWorked',
                       'PercentSalaryHike',
                       'PerformanceRating',
                       'TotalWorkingYears',
                       'TrainingTimesLastYear',
                       'YearsAtCompany',
                       'YearsInCurrentRole',
                       'YearsSinceLastPromotion',
                       'YearsWithCurrManager']
```

```python
# PATH
run_path = "/content/drive/MyDrive/0_AI_Projects/00_Project_Pulse/Attrition/runs/run_003"
json_file_path = "/content/drive/MyDrive/0_AI_Projects/00_Project_Pulse/Attrition/attrition_data.json"
df_path = "/content/drive/MyDrive/0_AI_Projects/00_Project_Pulse/Attrition/Employee_Attrition.csv"
```

```python
#Get problem context
prob_context = vi.load_problem_context(json_file_path)
```

```python
# Load df
import pandas as pd
df = pd.read_csv(df_path)
```

```python
for variable in numerical_variables:
  #Generate summary and save it
  summary = generate_summary(df, variable)
  save_summary(summary, run_path, variable)

  #Generate plots (Histogram and Boxplot)
  plot.generate_plots_num(df, variable, run_path)
```

```python
#AI-Interpretation of the variable
for variable in numerical_variables:
  variable_path = run_path + "/" + variable
  stream_markdown("**"+variable+"**")
  interpretation, latency = vi.variable_interpretation_engine(variable_path, prob_context)
  stream_markdown(interpretation) # removing markdown formatting
  print("Latency:", latency)
  stream_markdown("\n" + "-"*80 + "\n")
  print("\n")
  print("Press ENTER to continue...")
  input()
  print("\n")
```
